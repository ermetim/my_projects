```python
# Посмотрим все строки с пропущенными значениями в столбце 'year_of_release'
display(data[data['year_of_release'].isna()])
```

```python
# Заполним все пропуски в столбце 'year_of_release' нулем
data['year_of_release'] = data['year_of_release'].fillna(0)
```

```python
# Напишем функцию для поиска номера года в имени для заполнения пустых значений в столбце 'year_of_release'
def fill_year(empty_name):
    '''
    Функция ищет строки, в которых поле с годом выпуска равно нулю,
    и проверяет название игры на наличие года выпуска в названии.
    Если находит год в названии, то присваевает значение этого года столбцу 'year_of_release'.
    Годы взяты с 1980 по 2017 - выборка. 
    '''
    if empty_name['year_of_release'] == 0:
        for year in range(1980, 2017):
            if str(year) in empty_name['name']:
                print(year, end=',  ')
                return year
    return empty_name['year_of_release']
```

```python
# Запустим функцию по заполнению пропусков в столбце 'year_of_release'
data['year_of_release'] = data.apply(fill_year, axis=1)
```

```python
# Посчитаем долю пропущенных значений  в столбце 'year_of_release'
display(data.query('year_of_release == 0')['year_of_release'].count() / data['year_of_release'].count())
```

```python
#Удалим строки с пропущенными значениями в столбце 'year_of_release'
data = data.query('year_of_release != 0')
```

***3.2 Посмотрим, как менялись продажи по платформам. Выберем ТОП-10 платформ с наибольшими суммарными продажами и построим распределение по годам***

```python
# Найдем список топ-10 платформ с самыми высокими продажами по количеству копий. 
top_sales_platforms = (
    data
    .pivot_table(index='platform', values='total_sales', aggfunc='sum')
    .sort_values(by='total_sales', ascending=False)
    .reset_index()
    .head(10)
)
display(top_sales_platforms)
top_sales_platforms = top_sales_platforms['platform'].unique()
display(top_sales_platforms)
```

```python
platforms_by_years = (
    data
    .query('platform in @top_sales_platforms and year_of_release > 1993')
    .pivot_table(index=['year_of_release'], columns='platform', values='total_sales', aggfunc='sum')
)
display(platforms_by_years)
```

```python
plt.figure(figsize=(20,8))
plt.plot(platforms_by_years)
plt.legend(platforms_by_years.columns, loc='upper right')
plt.title('Распределение количества проданных копий на платформах')
plt.xlabel('Время, год')
plt.ylabel('Количество копий, шт')
plt.show()
```

Судя по графику можно заметить, что с появлением каждой платформы сначала идет резкий рост по количеству проданных копий, а затем идет падение продаж. По графику видно, что каждая платформа имеет характерный срок жизни и он составляет от 5 до 10 лет

***3.3 Выберем актуальный период***

Возьмем данные за соответствующий актуальный период. За актуальный период возьмем данные с 2005 до 2016 годы и проверим. Судя по графику распределения проданных копий можно наблюдать полный цикл развития платформ с максимальными продажами. 

```python
platforms = (
    data
    .query('2005 <= year_of_release <= 2015')
    .pivot_table(index=['year_of_release'], columns='platform', values='total_sales', aggfunc='sum')
)
display(platforms)
```

Возьмем актуальный период с 2013 года

```python
# Найдем список топ-5 платформ с самыми высокими продажами по количеству копий. 
top_actual_platforms = (
    data
    .query('2013 <= year_of_release <= 2015')
    .pivot_table(index='platform', values='total_sales', aggfunc='sum')
    .sort_values(by='total_sales', ascending=False)
    .reset_index()
    .head(5)
)
display(top_actual_platforms)
top_actual_platforms = top_actual_platforms['platform'].unique()
display(top_actual_platforms)
```

PS3, X360, 3DS	- продажи падают.

PS4 пришла на смену PS3.

XOne пришла на смену X360.

```python
# Выведем распределение продаж по регионам и выберем ТОП-5 платформ для каждого рагиона.
# Данные возмем с 2013 года включительно. Для этого напишем функцию.

def region_sales(index_col, region_sales):
    '''
    Функция выдает сводную таблицу в которую передаются 2 столбца. 
    Первый столбец будет индексом по которому производится группировка, 
    второй со значениями, которые суммируются.
    Также функция добавляет третий столбец, с долей от всего столбца.
    Данная функция нужна для составления портрета пользователя каждого региона
    '''
    region = (
        data
        .query('year_of_release >= 2013')
        .pivot_table(index=index_col, values=region_sales, aggfunc='sum')
        .sort_values(by=region_sales, ascending=False)
    )
    region['percentage'] = region[region_sales] / sum(region[region_sales])
    return region
```

***5.1. Гипотеза №1 - средние пользовательские рейтинги платформ Xbox One и PC одинаковые.***

За нулевую гипотезу примем нашу гипотезу №1 о том, что средний пользовательский рейтинг для платформы Xbox One и PC одинаковые.

За альтернативную гипотезу примем, что средний пользовательский рейтинг для платформы Xbox One и PC разные.
    
Для проверки нулевой гипотезы воспользуемся методом st.ttest_ind(), equal_var = False

```python
# Внесем переменную alpha_platform - критический уровень статистической значимости равный 5%
# Интервал берем с 2013 года
alpha_platform = .05

results_platform = st.ttest_ind(
    data.query('year_of_release >= 2013 and platform == "XOne" and user_score != 0')['user_score'], 
    data.query('year_of_release >= 2013 and platform == "PC" and user_score != 0')['user_score'],
    equal_var = False
)

print('p-значение: ', results_platform.pvalue)

if results_platform.pvalue < alpha_platform:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
display(data.query('year_of_release >= 2013 and platform == "XOne" and user_score != 0')['user_score'].mean())
display(data.query('year_of_release >= 2013 and platform == "PC" and user_score != 0')['user_score'].mean())
```

**Вывод:**

    На имеющихся данных, на уровне значимости 5% нулевую гипотезу можем отвергнуть. И на этом основании можно сказать, что средние пользовательские рейтинги платформ Xbox One и PC разные.

Для проверки гипотезы воспользуемся методом st.ttest_ind(), equal_var = False

    В Северной Америке платформы поделили пользователей равномерно и если в целом смотреть, то Playstation и Xbox имеют по 40%.
    
    
    4.2.	Определил для пользователя каждого региона (NA, EU, JP) самые популярные жанры (топ-5).
    По регионам Северная Америка и Европа жанры распределились приблизительно одинаково. На первом и втором местах Action и Shooter соответственно. Они забрали чуть больше 50% всех продаж. Далее идут Sports и Role-Playing которые забирют еще четверть всех продаж.
    А вот в регионе Япония картина обратная. Количество продаж на Action такое же как в Европе и Северной Америке (около 30%). Но данный жанр находится на втором мест. На первом же месте идет жанр Role-Playing и занимает 36%. То есть на 2 жанра приходится 65% всех продаж. На третьем месте жанр Misc - 7%. Далее идет жанты Fighting и Shooter по 5 %.
    
    4.3.	Определил влияет ли рейтинг ESRB на продажи в отдельном регионе (NA, EU, JP).      
    Также как и в жанрах для Северной Америки и Европы распределение в зависимости от рейтинга ESRB приблизительно одинаковое. То есть на первом месте с большим отрывом идут игры с рейтингом 'M' («Mature» — «Для взрослых») - ололо 38%, далее идет 'E' («Everyone» — «Для всех») - около 20%, далее идут 'E10+' («Everyone 10 and older» — «Для всех от 10 лет и старше») и 'T' («Teen» — «Подросткам») около 12% каждый. Даже доля неотмеченных значений порядка 20% для обоих регионов. Хотя это может быть потерей во время выгрузки. 

    По Японии сложно судить, так как доля нейзвестных позиций составляет 60%, поэтому какие-либо выводы сделать нельзя. 
    Единственное, что можно сказать наверняка, что даже при таком раскладе в Японии 'T' («Teen» — «Подросткам») имеет выше процент чем в Европе и Северной Америке. 

    Отсюда можно сделать вывод, что в Европе и Америке основной потребитель - это взрослое население. Когда как в Японии это возможно подростки, но это не точно.
